```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/manis/Downloads/Income - 2.csv')
```

```python
data = df.iloc[:,3:4].values
```

```python
plt.scatter(df['Age'],df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,7))
dendrogram = sch.dendrogram(sch.linkage(data,method = 'single'))
plt.title('Dendrogram')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'single')
cluster.fit_predict(data)
```

```python
cl = cluster.fit_predict(data)
from sklearn.metrics import silhouette_score
silhouette_score(data,cl)
```
